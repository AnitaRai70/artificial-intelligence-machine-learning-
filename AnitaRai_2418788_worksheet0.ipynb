Task1 - Unit Converter Program Step 1: Create Conversion Function

```python
def convert_units(conversion_type, value):
    """
    This function converts units of length, weight, and volume.

    Parameters:
    conversion_type (str): type of conversion
    value (float): value to convert

    Returns:
    float: converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084

    elif conversion_type == "ft_to_m":
        return value / 3.28084

    elif conversion_type == "kg_to_lbs":
        return value * 2.20462

    elif conversion_type == "lbs_to_kg":
        return value / 2.20462

    elif conversion_type == "l_to_gal":
        return value * 0.264172

    elif conversion_type == "gal_to_l":
        return value / 0.264172

    else:
        return "Invalid conversion type"
```

Step 2: Take User Input

```python
try:
    print("Choose Conversion Type")
    print("1: meters to feet")
    print("2: feet to meters")
    print("3: kg to pounds")
    print("4: pounds to kg")
    print("5: liters to gallons")
    print("6: gallons to liters")

    choice = int(input("Enter choice number: "))
    value = float(input("Enter value: "))

    if choice == 1:
        result = convert_units("m_to_ft", value)

    elif choice == 2:
        result = convert_units("ft_to_m", value)

    elif choice == 3:
        result = convert_units("kg_to_lbs", value)

    elif choice == 4:
        result = convert_units("lbs_to_kg", value)

    elif choice == 5:
        result = convert_units("l_to_gal", value)

    elif choice == 6:
        result = convert_units("gal_to_l", value)

    else:
        result = "Invalid choice"

    print("Converted value:", result)

except:
    print("Error: Please enter valid numbers")
```

TASK 2 - MAthematical Operations on List Step 1: create functions

```python
def find_sum(numbers):
    """Returns sum of numbers"""
    return sum(numbers)

def find_average(numbers):
    """Returns average"""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns maximum number"""
    return max(numbers)

def find_min(numbers):
    """Returns minimum number"""
    return min(numbers)
```

Step 2: User Input program

```python
try:
    nums = input("Enter numbers separated by space: ")
    numbers = list(map(float, nums.split()))

    print("Choose Operation")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    choice = int(input("Enter choice: "))

    if choice == 1:
        print("Sum =", find_sum(numbers))

    elif choice == 2:
        print("Average =", find_average(numbers))

    elif choice == 3:
        print("Maximum =", find_max(numbers))

    elif choice == 4:
        print("Minimum =", find_min(numbers))

    else:
        print("Invalid choice")

except:
    print("Invalid input")
```

TASK 3 - Extract every Other Element

```python
def extract_every_other(lst):
    return lst[::2]


my_list = [1,2,3,4,5,6]
result = extract_every_other(my_list)

print(result)
```

TASK 4 - Get Sublist

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]


my_list = [1,2,3,4,5,6]

print(get_sublist(my_list,2,4))
```

TASK 5 - Reverse List

```python
def reverse_list(lst):
    return lst[::-1]


my_list = [1,2,3,4,5]

print(reverse_list(my_list))
```

TASK 6 - Remove First and Last Element

```python
def remove_first_last(lst):
    return lst[1:-1]


my_list = [10,20,30,40,50]

print(remove_first_last(my_list))
```
