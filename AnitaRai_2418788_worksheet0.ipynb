Task1 - Unit Converter Program Step 1: Create Conversion Function

```python
def convert_units(conversion_type, value):
    """
    This function converts units of length, weight, and volume.

    Parameters:
    conversion_type (str): type of conversion
    value (float): value to convert

    Returns:
    float: converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084

    elif conversion_type == "ft_to_m":
        return value / 3.28084

    elif conversion_type == "kg_to_lbs":
        return value * 2.20462

    elif conversion_type == "lbs_to_kg":
        return value / 2.20462

    elif conversion_type == "l_to_gal":
        return value * 0.264172

    elif conversion_type == "gal_to_l":
        return value / 0.264172

    else:
        return "Invalid conversion type"
```

4.3 Exercise on Nested List:

1. Flatten a Nested List

```python
def flatten(lst):
    """
    Flattens a nested list (one level) into a single list.
    """
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(i)
        else:
            result.append(i)
    return result

# Example
nested_list = [[1, 2], [3, 4], [5]]
print("Flattened list:", flatten(nested_list))
```

2. Accessing Nested List Elements

```python
def access_nested_element(lst, indices):
    """
    Returns the element from a nested list using a list of indices.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element

# Example
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
print("Accessed element:", access_nested_element(lst, indices))
```

3. Sum of All Elements in a Nested List

```python
def sum_nested(lst):
    """
    Returns the sum of all numbers in a nested list, any depth.
    """
    total = 0
    for i in lst:
        if isinstance(i, list):
            total += sum_nested(i)
        else:
            total += i
    return total

# Example
nested_list = [[1, 2], [3, [4, 5]], 6]
print("Sum of all elements:", sum_nested(nested_list))
```

4. Remove Specific Element from a Nested List

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from a nested list.
    """
    result = []
    for i in lst:
        if isinstance(i, list):
            result.append(remove_element(i, elem))
        elif i != elem:
            result.append(i)
    return result

# Example
nested_list = [[1, 2], [3, 2], [4, 5]]
elem = 2
print("List after removal:", remove_element(nested_list, elem))
```

5. Find the Maximum Element in a Nested List

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list.
    """
    max_val = float('-inf')
    for i in lst:
        if isinstance(i, list):
            max_val = max(max_val, find_max(i))
        else:
            max_val = max(max_val, i)
    return max_val

# Example
nested_list = [[1, 2], [3, [4, 5]], 6]
print("Maximum element:", find_max(nested_list))
```

6. Count Occurrences of an Element in a Nested List

```python
def count_occurrences(lst, elem):
    """
    Counts how many times elem occurs in a nested list.
    """
    count = 0
    for i in lst:
        if isinstance(i, list):
            count += count_occurrences(i, elem)
        elif i == elem:
            count += 1
    return count

# Example
nested_list = [[1, 2], [2, 3], [2, 4]]
elem = 2
print("Occurrences of", elem, ":", count_occurrences(nested_list, elem))
```

7. Flatten a List of Lists of Lists

```python
def deep_flatten(lst):
    """
    Flattens a nested list of any depth into a single list.
    """
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)
    return result

# Example
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print("Deep flattened list:", deep_flatten(nested_list))
```

8. Nested List Average

```python
def average_nested(lst):
    """
    Returns the average of all elements in a nested list.
    """
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

# Example
nested_list = [[1, 2], [3, 4], [5, 6]]
print("Average:", average_nested(nested_list))
```

Problem 4: Matrix Operations


```python
import numpy as np

# ===============================
# Matrix Operations
# ===============================

# Given matrices
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# -------------------------------
# 1. Prove A * A^-1 = I
# -------------------------------
A_inv = np.linalg.inv(A)  # Compute inverse of A
identity_check = np.dot(A, A_inv)
print("\n1. A * A^-1 =\n", identity_check)

# Check if approximately equal to identity
print("Is approximately identity?", np.allclose(identity_check, np.eye(2)))

# -------------------------------
# 2. Prove AB != BA
# -------------------------------
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\n2. AB =\n", AB)
print("BA =\n", BA)
print("AB == BA?", np.array_equal(AB, BA))  # False, non-commutative

# -------------------------------
# 3. Prove (AB)^T = B^T * A^T
# -------------------------------
print("\n3. (AB)^T =\n", AB.T)
print("B^T * A^T =\n", np.dot(B.T, A.T))
print("Equal?", np.array_equal(AB.T, np.dot(B.T, A.T)))  # True

# -------------------------------
# 4. Solve system of linear equations
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9
# -------------------------------

# Coefficient matrix
A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

# Constants vector
B_vector = np.array([-1, -3, 9])

# Solve using inverse
X_inverse = np.dot(np.linalg.inv(A_matrix), B_vector)
print("\nSolution using inverse method [x, y, z]:", X_inverse)

# Solve using np.linalg.solve (more stable)
X_solve = np.linalg.solve(A_matrix, B_vector)
print("Solution using np.linalg.solve [x, y, z]:", X_solve)
```
